```python
# datnums = list(range(219, 225+1))
# datnums.remove(220)
# datnums = list(range(226, 234+1))
# for num in [228, 229]:
#     datnums.remove(num)
# datnums = [238, 243]
datnums = [244, 245]
dats = [get_dat(num) for num in datnums]

fig_isense, fig_isense_diff = go.Figure(), go.Figure()
figs = fig_isense, fig_isense_diff
fig_isense.update_layout(coloraxis_colorbar_title='Current /nA')
fig_isense.update_layout(coloraxis_colorbar_title='dI/dVg /nA')
for fig in figs:
    fig.update_layout(**default_layout,
                     title=f'Dats{datnums[0]}-{datnums[-1]}: Dot Tuning to weakly coupled regime',
                     xaxis_title=dats[0].Logs.x_label,
                     yaxis_title=dats[0].Logs.y_label)
for dat in dats:
    x = dat.Data.x
    y = dat.Data.y
    z = dat.Data.get_data('standard/i_sense')
    
    # I_sense
    z, x = U.resample_data(z, x, resample_x_only=True)
    fig_isense.add_trace(go.Heatmap(x=x, y=y, z=z, coloraxis='coloraxis'))
    
    # I_sense_diff
    z_diff = np.diff(z, axis=1)
    x_diff = U.get_matching_x(x, z_diff)
    fig_isense_diff.add_trace(go.Heatmap(x=x_diff, y=y, z=z_diff, coloraxis='coloraxis'))
    
for fig in figs:
    fig.show(config=default_config)
    
```
